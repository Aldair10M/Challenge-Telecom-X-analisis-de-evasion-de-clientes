```python
import pandas as pd;
```

#ðŸ“Œ ExtracciÃ³n

```python
# Paso 1: Cargar los datos desde la API
api = 'https://raw.githubusercontent.com/Aldair10M/Challenge-Telecom-X-an-lisis-de-evasi-n-de-clientes/refs/heads/main/TelecomX_Data.json';
datos = pd.read_json(api);
datos
```

```python
# Paso 2: Normalizar los datos
datos_normalizados = pd.json_normalize(datos.to_dict(orient='records'), sep='_')
# Paso 3 (opcional): Visualizar
datos_normalizados
```

#ðŸ”§ TransformaciÃ³n

```python
# A travÃ©s de info() vemo un resumen general del DataFrame: nÃºmero de columnas, valores nulos y tipos de datos
datos_normalizados.info()
```

```python
# A travÃ©s de dtypes vemos sÃ³lo los tipos de datos por columna
print("\nTipos de datos:")
print(datos_normalizados.dtypes)
```

```python
# Convertimos la columna 'account_Charges_Total' a numÃ©rico, ya que es de tipo object (String)
datos_normalizados['account_Charges_Total'] = pd.to_numeric(datos_normalizados['account_Charges_Total'], errors='coerce')
```

```python
# Re-verificamos los tipos
print("\nTipos de datos:")
print(datos_normalizados.dtypes)
```

```python
# 1. Verificamos los valores nulos
print("Valores nulos por columna:")
print(datos_normalizados.isnull().sum())
```

```python
# 2. Verificamos las filas duplicados
duplicados = datos_normalizados.duplicated().sum()
print(f"NÃºmero de filas duplicadas: {duplicados}")

## si es que existen filas duplicadas:
# datos_normalizados = datos_normalizados.drop_duplicates()
```

```python
# 3. Verificamos y limpiamos los valores inconsistentes en categorÃ­as
print("\nValores Ãºnicos en 'Churn':", datos_normalizados['Churn'].unique())
print("Valores Ãºnicos en 'customer_gender':", datos_normalizados['customer_gender'].unique())
print("Valores Ãºnicos en 'customer_Partner':", datos_normalizados['customer_Partner'].unique())
print("Valores Ãºnicos en 'phone_PhoneService':", datos_normalizados['phone_PhoneService'].unique())
print("Valores Ãºnicos en 'account_Contract':", datos_normalizados['account_Contract'].unique())
print("Valores Ãºnicos en 'account_PaperlessBilling':", datos_normalizados['account_PaperlessBilling'].unique())
```

```python
# Eliminar filas con 'Churn' vacÃ­o (string vacÃ­o)
datos_normalizados = datos_normalizados[datos_normalizados['Churn'] != '']
print("Valores Ãºnicos en 'Churn' (post-limpieza):", datos_normalizados['Churn'].unique())
```

```python
print("\nValores Ãºnicos en 'Churn':", datos_normalizados['Churn'].unique())
print("Valores Ãºnicos en 'customer_gender':", datos_normalizados['customer_gender'].unique())
print("Valores Ãºnicos en 'customer_Partner':", datos_normalizados['customer_Partner'].unique())
print("Valores Ãºnicos en 'phone_PhoneService':", datos_normalizados['phone_PhoneService'].unique())
print("Valores Ãºnicos en 'account_Contract':", datos_normalizados['account_Contract'].unique())
print("Valores Ãºnicos en 'account_PaperlessBilling':", datos_normalizados['account_PaperlessBilling'].unique())
```

```python
# 4. Eliminamos filas con 'account_Charges_Total' nulos
datos_normalizados = datos_normalizados.dropna(subset=['account_Charges_Total'])
```

```python
print("Valores nulos por columna:")
print(datos_normalizados.isnull().sum())
```

```python
# 1. Crear la columna "Cuentas_Diarias"
datos_normalizados['Cuentas_Diarias'] = datos_normalizados['account_Charges_Monthly'] / 30

# 2. Convertir "yes"/"no" a 1 y 0
cols_binarias = [col for col in datos_normalizados.columns if datos_normalizados[col].isin(['yes', 'no']).all()]
for col in cols_binarias:
    datos_normalizados[col] = datos_normalizados[col].map({'yes': 1, 'no': 0})

# 3. Renombrar algunas columnas para facilitar la comprensiÃ³n
datos_normalizados = datos_normalizados.rename(columns={
    'Churn': 'Evasion',
    'customer_gender': 'Genero',
    'customer_SeniorCitizen': 'Adulto_Mayor',
    'customer_Partner': 'Tiene_Pareja',
    'customer_Dependents': 'Tiene_Dependientes',
    'customer_tenure': 'Meses_Cliente',
    'account_Charges_Monthly': 'Facturacion_Mensual',
    'account_Charges_Total': 'Facturacion_Total',
    'account_Contract': 'Tipo_Contrato'
})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# ConfiguraciÃ³n de estilo
sns.set(style="whitegrid")

# Conteo de valores
plt.figure(figsize=(6,4))
sns.countplot(x='Evasion', data=datos_normalizados, palette='Set2')

plt.title('DistribuciÃ³n de EvasiÃ³n de Clientes (Churn)')
plt.xlabel('EvasiÃ³n (0 = No, 1 = SÃ­)')
plt.ylabel('NÃºmero de Clientes')
plt.xticks([0, 1], ['No se fue', 'SÃ­ se fue'])
plt.show()
```

```python
# Pie chart
evasion_counts = datos_normalizados['Evasion'].value_counts()
labels = ['No se fue', 'SÃ­ se fue']
colors = ['skyblue', 'salmon']

plt.figure(figsize=(6,6))
plt.pie(evasion_counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('ProporciÃ³n de EvasiÃ³n de Clientes')
plt.axis('equal')
plt.show()
```

```python
# EvasiÃ³n por GÃ©nero
plt.figure(figsize=(6,4))
sns.countplot(x='Genero', hue='Evasion', data=datos_normalizados, palette='Set2')
plt.title('EvasiÃ³n por GÃ©nero')
plt.xlabel('GÃ©nero')
plt.ylabel('Cantidad de clientes')
plt.legend(title='EvasiÃ³n', labels=['No', 'SÃ­'])
plt.show()

# EvasiÃ³n por Tipo de Contrato
plt.figure(figsize=(7,4))
sns.countplot(x='Tipo_Contrato', hue='Evasion', data=datos_normalizados, palette='Set1')
plt.title('EvasiÃ³n por Tipo de Contrato')
plt.xlabel('Tipo de Contrato')
plt.ylabel('Cantidad de clientes')
plt.legend(title='EvasiÃ³n', labels=['No', 'SÃ­'])
plt.xticks(rotation=15)
plt.show()

# EvasiÃ³n por MÃ©todo de Pago
plt.figure(figsize=(10,4))
sns.countplot(x='account_PaymentMethod', hue='Evasion', data=datos_normalizados, palette='Set3')
plt.title('EvasiÃ³n por MÃ©todo de Pago')
plt.xlabel('MÃ©todo de Pago')
plt.ylabel('Cantidad de clientes')
plt.legend(title='EvasiÃ³n', labels=['No', 'SÃ­'])
plt.xticks(rotation=45)
plt.show()
```

```python
# Tiempo de permanencia vs EvasiÃ³n
plt.figure(figsize=(6,4))
sns.boxplot(x='Evasion', y='Meses_Cliente', data=datos_normalizados, palette='Pastel1')
plt.title('Meses como Cliente vs EvasiÃ³n')
plt.xlabel('EvasiÃ³n (0 = No, 1 = SÃ­)')
plt.ylabel('Meses como Cliente')
plt.show()

# Total Facturado vs EvasiÃ³n
plt.figure(figsize=(6,4))
sns.boxplot(x='Evasion', y='Facturacion_Total', data=datos_normalizados, palette='Pastel2')
plt.title('FacturaciÃ³n Total vs EvasiÃ³n')
plt.xlabel('EvasiÃ³n (0 = No, 1 = SÃ­)')
plt.ylabel('FacturaciÃ³n Total')
plt.show()
```

## ðŸ”¹ IntroducciÃ³n
Telecom X enfrenta una alta tasa de cancelaciÃ³n de servicios por parte de sus clientes. El objetivo de este anÃ¡lisis fue comprender mejor los factores que influyen en la evasiÃ³n de clientes (Churn), con el fin de aportar informaciÃ³n valiosa para estrategias de retenciÃ³n.

## ðŸ”¹ Limpieza y Tratamiento de Datos
- Se importaron datos desde una API en formato JSON.
- Se normalizÃ³ la estructura jerÃ¡rquica de los datos para convertirla en un DataFrame plano.
- Se revisaron los tipos de datos y se convirtieron valores mal tipados como `account_Charges_Total`.
- Se eliminaron valores nulos e inconsistencias, como cadenas vacÃ­as en la variable 'Churn'.
- Se transformaron respuestas categÃ³ricas ("yes"/"no") a valores binarios (1/0).
- Se creÃ³ una nueva columna `Cuentas_Diarias` a partir de la facturaciÃ³n mensual.

## ðŸ”¹ AnÃ¡lisis Exploratorio de Datos
- Se realizÃ³ un anÃ¡lisis descriptivo para conocer medias, medianas y desviaciones estÃ¡ndar.
- Se observÃ³ que el 26% de los clientes abandonaron el servicio (`Churn` = "yes").
- Visualizaciones mostraron:
  - Mayor evasiÃ³n en contratos `month-to-month`.
  - Usuarios sin pareja o dependientes tienen ligeramente mayor evasiÃ³n.
  - MÃ©todos de pago como `electronic check` se asocian con mÃ¡s evasiÃ³n.
  - Clientes con menor tiempo de permanencia (`Meses_Cliente`) y menor gasto (`FacturaciÃ³n_Total`) tienden a cancelar mÃ¡s.

## ðŸ”¹ Conclusiones e Insights
- El tipo de contrato es un fuerte predictor de evasiÃ³n: los contratos mensuales tienen mÃ¡s riesgo de churn.
- Los clientes con menor historial y facturaciÃ³n total son mÃ¡s propensos a cancelar.
- Variables como `PaperlessBilling` o `MÃ©todo de Pago` tambiÃ©n tienen relaciÃ³n con el churn.
- Factores demogrÃ¡ficos como gÃ©nero o tener pareja tienen influencia moderada.

## ðŸ”¹ Recomendaciones
- Incentivar contratos de mayor duraciÃ³n con promociones o descuentos.
- Identificar y retener clientes de bajo tenure con campaÃ±as personalizadas.
- Mejorar la experiencia de facturaciÃ³n electrÃ³nica, dado que se asocia a mayor evasiÃ³n.
- Segmentar clientes por riesgo y aplicar modelos predictivos para acciones proactivas de retenciÃ³n.
